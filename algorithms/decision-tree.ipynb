# **Decision Tree** 
* A Decision Tree is a tree-like model used for classification and regression.
* It splits data into branches based on feature values until it reaches a decision (final output).

![](https://raw.githubusercontent.com/prasannashrestha011/ImageRepository/main/uploads/DecisonTree.png)

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data_set=load_breast_cancer()
X=data_set['data']
y=data_set['target']
tree=DecisionTreeClassifier(max_depth=4,random_state=0)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
tree.fit(X,y)
```

```python
print("Training score:{:.2f}".format(tree.score(X_train,y_train)))
print("Testing score:{:.2f}".format(tree.score(X_test,y_test)))
```

```python

from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(25, 14))
plot_tree(tree, feature_names=data_set.feature_names, class_names=data_set.target_names, filled=True,fontsize=12)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
```

```python
iris_dataset=load_iris()
X,y=iris_dataset['data'],iris_dataset['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)

tree=DecisionTreeClassifier(random_state=42)
```

```python
tree.fit(X_train,y_train)
```

```python
print("Training score {:.2f}".format(tree.score(X_train,y_train)))
print("Test score {:.2f}".format(tree.score(X_test,y_test)))
```

```python
plt.figure(figsize=(28,19))
plot_tree(tree,feature_names=iris_dataset.feature_names,class_names=iris_dataset.target_names,
         filled=True,fontsize=19)
```
