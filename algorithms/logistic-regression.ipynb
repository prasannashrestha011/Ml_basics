# Logisitic Regression 
Logisitic Regression algorithm is used for classification task.It computes the probabilty of instance belongs to given task or not.

Logistic regression is used for binary classification where we use sigmoid function, that takes input as independent variables and produces a probability value between 0 and 1.

In sigmoid function, it uses z=mx+b formula as base and gives the probabilty of the instance to find its belonging class

![](https://raw.githubusercontent.com/prasannashrestha011/ImageRepository/main/uploads/1694183259538.png)

For instance, in the datasets with class 0 and 1, if the probabilty of data P(n)=0.35 <0.5, it  belongs to class 0 whereas if P(n)=0.67 > 0.5, it  belongs to the class 1
