```python
file_path="../input/email-spam-classification-dataset/combined_data.csv"
df=pd.read_csv(file_path)
pd.DataFrame(df[:5])
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)  
X_tfidf = vectorizer.fit_transform(df['text'])  # Convert emails to numerical format
y_input=df.label
```

```python

## spliting the dataset
X_train,X_test,y_train,y_test=train_test_split(X_tfidf,y_input,test_size=0.25,random_state=66)
```

```python
model = LogisticRegression(solver='saga', max_iter=2000, class_weight='balanced')
model.fit(X_train,y_train)
```

```python
def predict_email(text):
    text_tfidf=vectorizer.transform([text])
    prediction=model.predict(text_tfidf)
    return "Spam" if prediction[0]==1 else "Not spam"
```

```python
custom_text=str(input("Enter your email description: \n "))
print("\n")
predict_email(custom_text)
```
