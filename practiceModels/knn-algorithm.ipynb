```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
feature=iris.data
target=iris.target
X_train,X_test,y_train,y_test=train_test_split(feature,target,test_size=0.25,random_state=42)
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
# training the model with 
knn.fit(X_train,y_train)
```

```python
X_input=np.array([[2.2,5,1.1,3]])
prediction=knn.predict(X_input)
iris.target_names[prediction]
```

```python
# finding model accuracy by using test data (25%)
y_predict=knn.predict(X_test)
# comparing predicted output with y_test 
accuracy=accuracy_score(y_test,y_predict)
accuracy
```
