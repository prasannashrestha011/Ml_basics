# Random Forest
In random forest, its has multiple decision tree. Each tree will pick random subset of feature data
for training and give a prediction.If the task is classification , majority voted class get selected whereas in 
regression algorithm will take a average of predicited value given by each tree

**Advantage** 
* Avoid overfitting
* Increase prediction accuracy

**Disadvantage**
* Time consuming for larger dataset
* Interpretability is difficut

**Determining feature importance** 
 Random forest gives feature importance by measuring how each feature help model in     making good prediction. 
 Two method 
 * Gini importance->checks how each feature split the data correctly in decision tree.
                     if feature is good and reduce uncertanity, it gets higher
                     importance score.
 * Permutation importance ->In this method, it random suffles the each features values
                            and if the model accuracy drops significantly then suffled                             feature will be marked as important

```python
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier

from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X,y=iris.data,iris.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
rf=RandomForestClassifier().fit(X_train,y_train)
```

```python

print("Training score :{:.2f}".format(rf.score(X_train,y_train)))
print("Testing score :{:.2f}".format(rf.score(X_test,y_test)))
```

```python
feature_names=iris.feature_names
feature_importance=rf.feature_importances_
plt.bar(feature_names,feature_importance,width=0.5)
plt.xlabel("Feature importances")
```
